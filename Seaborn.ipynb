1.Seaborn Plots and Their Uses
Line Plot:

Use: Used to visualize the trend of a numerical variable over a continuous interval or time period. It is helpful for time series data analysis.
Example: Showing changes in stock prices over time.
Box Plot:

Use: Displays the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps identify outliers and the spread of the data.
Example: Comparing test scores across different classes.
Histogram:

Use: Visualizes the distribution of a single numerical variable by dividing the data into bins and counting the number of observations in each bin.
Example: Showing the distribution of house prices in a city.
Pair Plot:

Use: Creates a matrix of scatter plots for multiple numerical variables, along with histograms or KDE plots on the diagonal. It helps in understanding the relationships between variables.
Example: Exploring the pairwise relationships between features in the Iris dataset.
Heatmap:

Use: Displays data in a matrix format using colors to represent the intensity or value of each cell. It is useful for showing correlations or frequencies.
Example: Visualizing the correlation matrix of different features in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.title("FMRI Signal Over Time by Event and Region")
plt.show()
```

```python
# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Box plot for age
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Passenger Class")

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Histogram for price with hue for cut
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")

# Display the plot
plt.title("Price Distribution of Diamonds by Cut")
plt.show()

```
